```python
import warnings
import datetime
import numpy as np
import pandas as pd
import sklearn
import matplotlib.pyplot as plt

warnings.filterwarnings('ignore')
%matplotlib inline
plt.rcParams['figure.figsize'] = [10, 5]
plt.rcParams['figure.dpi'] = 200
RANDOM_SEED = 6324
np.random.seed(RANDOM_SEED)
```

### Data processing

```python
df = pd.read_csv('nasdaq_big.csv')
dates = np.array([datetime.datetime.strptime(date,"%Y-%m-%d") for date in df["Date"].tolist()])
y = df['y'].to_numpy()
x = df[['x1','x2','x3','x4','x5']].to_numpy()
t = np.arange(len(y))
ci_multiplier = 0.4
```

```python
# Split
i = int(len(y)*0.8)
dates_train = dates[:i]
dates_test = dates[i:]
y_train = y[:i]
y_test = y[i:]
x_train = x[:i]
x_test = x[i:]
t_train = t[:i]
t_test = t[i:]
```

```python
t_train = t_train.reshape(len(t_train),1)
y_train = y_train.reshape(len(y_train),1)
```

### Gaussian Process

```python
from sklearn.gaussian_process import GaussianProcessRegressor
from sklearn.gaussian_process.kernels import WhiteKernel, ConstantKernel, RBF#ExpSineSquared

k0 = WhiteKernel(noise_level=3**2, noise_level_bounds=(1**2, 5**2))
k1 = ConstantKernel(constant_value=np.mean(y)) * RBF(length_scale=10)
kernel = k0 + k1

gp1 = GaussianProcessRegressor(kernel=kernel,
                               n_restarts_optimizer=10,
                               normalize_y=True,
                               copy_X_train=True,
                               alpha=0.0)
```

```python
gp1.fit(x_train, y_train)
```

```python
y_pred, y_std = gp1.predict(x_test, return_std=True)
```

```python
data_df = pd.DataFrame()
data_df['t'] = t_test
data_df['y1'] = y_test
data_df['y_pred'] = y_pred
data_df['y_std'] = y_std
data_df['y_pred_lwr'] = data_df['y_pred'] - 0.4*data_df['y_std']
data_df['y_pred_upr'] = data_df['y_pred'] + 0.4*data_df['y_std']
```

```python
upper_bound_unscaled = np.array(data_df['y_pred_upr'])
lower_bound_unscaled = np.array(data_df['y_pred_lwr'])
under_upper = upper_bound_unscaled > y_test
over_lower = lower_bound_unscaled < y_test
total = (under_upper == over_lower)

print("{} our predictions are in our confidence interval".format(np.mean(total)))
```

```python
from sklearn.metrics import mean_squared_error
import math
y_actual = y_test
y_predicted = y_pred
 
MSE = mean_squared_error(y_actual, y_predicted)
 
RMSE = math.sqrt(MSE)
print("Root Mean Square Error:\n")
print(RMSE)
```

```python
params = {"ytick.color" : "black",
          "xtick.color" : "black",
          "axes.labelcolor" : "black",
          "axes.edgecolor" : "black"}

plt.rcParams.update(params)

plt.title("NASDAQ Index Price - GPR")

plt.plot(df.index,
         y,
         color='black',
         label="Real")

plt.plot(t_test,
         y_pred,
         label="Prediction",
         color="red")

locs, labels = plt.xticks()
plt.xticks([0,460,920,1380,1840,2300],['2011','2013','2015','2017','2019','2021'])

plt.fill_between(x=t_test,
                 y1=upper_bound_unscaled,
                 y2=lower_bound_unscaled,
                 facecolor='green',
                 label="Credible interval",
                 alpha=0.5)

plt.legend(loc=2)
```
