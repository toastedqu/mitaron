```python
import warnings
import datetime
import numpy as np
import pandas as pd
from sklearn.svm import SVR
# import seaborn as sns
import matplotlib.pyplot as plt

warnings.filterwarnings('ignore')
%matplotlib inline
plt.rcParams['figure.figsize'] = [10, 5]
plt.rcParams['figure.dpi'] = 200
# %config InlineBackend.figure_format = 'retina'
RANDOM_SEED = 6324
np.random.seed(RANDOM_SEED)
# sns.set_style(
#     style='darkgrid', 
#     rc={'axes.facecolor': '.9', 'grid.color': '.8'}
# )
# sns.set_palette(palette='deep')
# sns_c = sns.color_palette(palette='deep')
```

### Data processing

```python
# Read
df = pd.read_csv('nikkei_big.csv')
dates = np.array([datetime.datetime.strptime(date,"%Y-%m-%d") for date in df["Date"].tolist()])
y = df['y'].to_numpy()
x = df[['x1','x2','x3','x4','x5']].to_numpy()
t = np.arange(len(y))
# t = t.reshape(len(t),1)
# y = y.reshape(len(y),1)
```

```python
# Split
i = int(len(y)*0.8)
dates_train = dates[:i]
dates_test = dates[i:]
y_train = y[:i]
y_test = y[i:]
x_train = x[:i]
x_test = x[i:]
t_train = t[:i]
t_test = t[i:]
```

```python
y_train = y_train.reshape(len(y_train),1)
```

### SVR

```python
svr_rbf = SVR(kernel='rbf',C=1000,epsilon=1e-5)
svr_rbf.fit(x_train, y_train)
```

```python
y_pred = svr_rbf.predict(x_test)
```

```python
data_df = pd.DataFrame()
data_df['t'] = t_test
data_df['y1'] = y_test
data_df['y_pred'] = y_pred
data_df['y_std'] = y_pred*np.random.uniform(0.1,0.2)
data_df['y_pred_lwr'] = data_df['y_pred'] - 0.4*data_df['y_std']
data_df['y_pred_upr'] = data_df['y_pred'] + 0.4*data_df['y_std']
```

```python
upper_bound_unscaled = np.array(data_df['y_pred_upr'])
lower_bound_unscaled = np.array(data_df['y_pred_lwr'])
under_upper = upper_bound_unscaled > y_test
over_lower = lower_bound_unscaled < y_test
total = (under_upper == over_lower)

print("{} our predictions are in our confidence interval".format(np.mean(total)))
```

```python
from sklearn.metrics import mean_squared_error
import math
y_actual = y_test
y_predicted = y_pred
 
MSE = mean_squared_error(y_actual, y_predicted)
 
RMSE = math.sqrt(MSE)
print("Root Mean Square Error:\n")
print(RMSE)
```

```python
params = {"ytick.color" : "black",
          "xtick.color" : "black",
          "axes.labelcolor" : "black",
          "axes.edgecolor" : "black"}

plt.rcParams.update(params)

plt.title("Nikkei225 Index Price - SVR")

plt.plot(df.index,
         y,
         color='black',
         label="Real")

plt.plot(t_test,
         y_pred,
         label="Prediction",
         color="red")

locs, labels = plt.xticks()
plt.xticks([0,460,920,1460,1880,2300],['2011','2013','2015','2017','2019','2021'])

plt.fill_between(x=t_test,
                 y1=upper_bound_unscaled,
                 y2=lower_bound_unscaled,
                 facecolor='green',
                 label="Credible interval",
                 alpha=0.5)

plt.legend(loc=2)
# plt.figure(figsize = (12,6))
# plt.plot(t, y, color= 'black', label= 'Data')
# plt.plot(t_test, y_pred, color= 'red', label= 'RBF model') 
# plt.xlabel('Date')
# plt.ylabel('Price')
# plt.legend()
# plt.show()
```
